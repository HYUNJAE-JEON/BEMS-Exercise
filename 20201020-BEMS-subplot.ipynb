```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpld3 import enable_notebook
enable_notebook()
X = np.random.normal(0, 1, (100, 3))
color = np.random.random(100)
size = 500 * np.random.random(100)
plt.scatter(X[:,0], X[:,1], c=color,
           s=size, alpha=0.5, linewidth=2)
plt.grid(color='lightgray', alpha=0.7)

```

```python
fig, ax = plt.subplots(3, 3, figsize=(10,10), sharex=True, sharey=True)
fig.subplots_adjust(hspace=0.3)
#간격을 조정하겠다. adjust
X[::2,2] += 3
for i in range(3):
    for j in range(3):
        ax[i,j].scatter(X[:,i], X[:,j], c=color, s=.1*size, alpha=0.5, linewidths=2)
        ax[i,j].grid(color='lightgray', alpha=0.7)


#여러 개를 만들 때는 subplots를 쓰면 된다.
#이거는 굉장히 중요하다.
#line 2개를 조절할 때는 linewidths, 크기를 조절할 때는 s를 조절. 투명도를 조절할 때는 alpha, subplots에 리턴을 2개 받아서 조절하는 구나.
```
