```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
boston = load_boston()
dfX_boston = pd.DataFrame(boston.data, columns=boston.feature_names) #column = feature로 본다.
dfy_boston = pd.DataFrame(boston.target, columns = ["MDEV"])
df_boston0 = pd.concat([dfX_boston, dfy_boston], axis=1)
df_boston0.head()

#numpy는 데이터 합치는게 concatenate, pandas는 concat
```

```python
import seaborn as sns
df_boston0.describe()
cols = ["LSTAT", "NOX", "RM", "MDEV"]
sns.pairplot(df_boston0[cols])
plt.show()
```

```python
import numpy as np
from sklearn.datasets import load_boston
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
import matplotlib
boston = load_boston()
x = boston.data
y = boston.target
lr = LinearRegression()
lr.fit(x,y)
y_lrp = lr.predict(x)
rmse = np.sqrt(mean_squared_error(y, y_lrp))
print(f'RMSE: {rmse}')
plt.plot([0,50],[0,50], '-', color=(.9, .3, .3), lw=4)
plt.scatter(y_lrp, y)
import matplotlib
matplotlib.rc('font', family='HYGothic-Medium')
plt.xlabel('예측가격')
plt.ylabel('실제가격')
plt.savefig('보스턴집값')
```
